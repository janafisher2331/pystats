# Hypothesis Testing

### Test likelihood that value of population mean will equal zero.

```python
from scipy import stats
```

```python
stats.ttest_1samp(data['VIQ'], 0)
```

### p-value indicates this is not true.

### Use 2-sample t test to find significance between populations.

```python
female_viq = data[data['Gender'] == 'Female']['VIQ']
```

```python
male_viq = data[data['Gender'] == 'Male']['VIQ']
```

```python
stats.ttest_ind(female_viq, male_viq) 
```

### Test significance between FSIQ and PIQ.

```python
stats.ttest_ind(data['FSIQ'], data['PIQ'])
```

### Use repeated measures test for significance between FSIQ and PIQ.

```python
stats.ttest_rel(data['FSIQ'], data['PIQ'])
```

### Repeated measures test is equivalent to 1-sample test on the difference.

```python
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0)
```

### Use Wilcoxon signed-rank test to ignore Gaussian errors.

```python
stats.wilcoxon(data['FSIQ'], data['PIQ'])
```

### First, generate simulated data according to model.

```python
import numpy as np
```

```python
data = pandas.DataFrame({'x': x, 'y': y})
```

### Then specify OLS model and fit.

```python
from statsmodels.formula.api import ols
```

```python
model = ols("y ~ x", data).fit()
```

### View statistics from fit.
